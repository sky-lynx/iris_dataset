```python
df = pd.read_csv('iris.data', header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
df
```

```python
df.describe()
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize = (15,12))
sns.distplot(df.sepal_length, bins=21, ax=axes[0,0])
sns.distplot(df.sepal_width, bins = 21, ax=axes[0,1])
sns.distplot(df.petal_length, bins = 20, ax=axes[1,0])
sns.distplot(df.petal_width, bins = 20, ax=axes[1,1])
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize = (15,12))
sns.boxplot(x = 'class', y = 'sepal_length', data = df, ax=axes[0,0])
sns.boxplot(x = 'class', y = 'sepal_width', data = df, ax=axes[0,1])
sns.boxplot(x = 'class', y = 'petal_length', data = df, ax=axes[1,0])
sns.boxplot(x = 'class', y = 'petal_width', data = df, ax=axes[1,1])
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize = (15,12))
sns.violinplot(x = 'class', y = 'sepal_length', data = df, ax=axes[0,0])
sns.violinplot(x = 'class', y = 'sepal_width', data = df, ax=axes[0,1])
sns.violinplot(x = 'class', y = 'petal_length', data = df, ax=axes[1,0])
sns.violinplot(x = 'class', y = 'petal_width', data = df, ax=axes[1,1])
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(), center=0)
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize = (15,12))
sns.regplot(x = df.sepal_length, y = df.sepal_width, ax=axes[0,0])
sns.regplot(x = df.sepal_length, y = df.petal_length, ax=axes[0,1])
sns.regplot(x = df.sepal_length, y = df.petal_width, ax=axes[1,0])
```
